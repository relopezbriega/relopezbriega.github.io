## Introducción

El concepto de *[infinito](https://es.wikipedia.org/wiki/Infinito)* ha obsesionado a la mente humana por miles de años. La idea de que las cosas pueden seguir y seguir para siempre, que pueden no tener ni principio ni final, ni centro, ni límites; desafía a la intuición. Científicos, filósofos y teólogos por igual, han tratado de entenderlo, de cortar su tamaño, de averiguar su forma y dimensión; y en última instancia, de decidir si es un concepto bienvenido en nuestras descripciones del Universo que nos rodea.
A pesar de todas las dificultades que puede traer el lidiar con el *[infinito](https://es.wikipedia.org/wiki/Infinito)*, éste sigue siendo un tema fascinante. Se encuentra en el corazón de todo tipo de preguntas fundamentales del hombre. ¿Se puede vivir para siempre? ¿El universo tiene un final? ¿Tuvo un comienzo? ¿Tiene el universo un "borde" o simplemente no existen límites a su tamaño? Aunque es fácil pensar en las listas de números o secuencias de 'tics' de un reloj que continúan para siempre, hay otros tipos de *[infinito](https://es.wikipedia.org/wiki/Infinito)* que parecen ser más difíciles de concebir.

## ¿Qué es el infinito?

Generalmente existe la tendencia de pensar en el *[infinito](https://es.wikipedia.org/wiki/Infinito)* simplemente como un número muy grande, sólo un poco más grande que el mayor número que se pueda imaginar, siempre fuera de su alcance; pero **el *[infinito](https://es.wikipedia.org/wiki/Infinito)* no es un número**; sino que es un concepto,  la idea de ***algo que no tiene fin***. 

## El infinito en la matemáticas

Si bien el concepto de *[infinito](https://es.wikipedia.org/wiki/Infinito)* parece ser un tema elusivo y complejo, aun así no a escapado a su estudio por parte de las [matemáticas](https://es.wikipedia.org/wiki/Matem%C3%A1ticas). Su estudio ya comenzó en la antigua Grecia, quienes con [teoremas](https://es.wikipedia.org/wiki/Teorema) tales como que la cantidad de [números primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) no tiene límites, se enfrentaron ante la perspectiva de lo *[infinito](https://es.wikipedia.org/wiki/Infinito)*. [Aristóteles](https://es.wikipedia.org/wiki/Arist%C3%B3teles) evitó la actualidad del *[infinito](https://es.wikipedia.org/wiki/Infinito)* mediante la definición de una *infinidad potencial*, de forma de permitir que éstos [teoremas](https://es.wikipedia.org/wiki/Teorema) continúen como válidos. Según esta distinción hecha por [Aristóteles](https://es.wikipedia.org/wiki/Arist%C3%B3teles) entre el [infinito actual y el infito potencial](https://es.wikipedia.org/wiki/Infinito_potencial_e_infinito_actual), los [números enteros](https://es.wikipedia.org/wiki/N%C3%BAmero_entero) son potencialmente *[infinitos](https://es.wikipedia.org/wiki/Infinito)* porque siempre puede sumarse uno para conseguir un número mayor, pero el conjunto *[infinito](https://es.wikipedia.org/wiki/Infinito)* de [números enteros](https://es.wikipedia.org/wiki/N%C3%BAmero_entero) como tal, es decir el *[infinito](https://es.wikipedia.org/wiki/Infinito)* actual de números enteros, no existe. Según los griegos el *[infinito](https://es.wikipedia.org/wiki/Infinito)* actual nunca puede existir, sino que solo puede existir el *[infinito](https://es.wikipedia.org/wiki/Infinito)* potencial; como reflejo del pensamiento griego, [Aristóteles](https://es.wikipedia.org/wiki/Arist%C3%B3teles) concluye que el *[infinito](https://es.wikipedia.org/wiki/Infinito)* es imperfecto, inacabado e impensable. Esta definición de un *[infinito](https://es.wikipedia.org/wiki/Infinito)* potencial y no real, funcionó y satisfizo a los matemáticos y filósofos por casi dos milenios.

### Notación del infinito

El símbolo $\infty $ que representa al *[infinito](https://es.wikipedia.org/wiki/Infinito)* se lo debemos al matemático ingles [John Wallis](https://es.wikipedia.org/wiki/John_Wallis), quien fue un precursor del [cálculo infinitesimal](https://es.wikipedia.org/wiki/Infinitesimal) e introdujo la notación $\infty$, que representa una curva sin fin, en el siglo diecisiete.

### En camino hacia un mejor entendimiento del infinito

A medida que se continuaba explorando el concepto del *[infinito](https://es.wikipedia.org/wiki/Infinito)*, nuevas [paradojas](https://es.wikipedia.org/wiki/Paradoja) continuaban apareciendo. En la edad media se entendía que un círculo más grande debería contener más puntos que un círculo más pequeño, sin embargo se puede encontrar una correspondencia de *uno a uno* entre los puntos de ambos círculos, como lo demuestra la siguiente figura:

<img alt="Paradoja de los círculos" title="Paradoja de los círculos" src="http://relopezbriega.github.io/images/paradoja-circulos.png
" width="300px" height="300px">

En 1600, [Galileo](https://es.wikipedia.org/wiki/Galileo_Galilei) entendió que el problema era el estar utilizando un razonamiento finito sobre cosas infinitas. Postuló que "Es un error hablar de cantidades infinitas como siendo la mayor o menor o igual que a la otra" y afirmó que el *[infinito](https://es.wikipedia.org/wiki/Infinito)* no es un noción inconsistente, sino que obedece a reglas diferentes.

Otro de los temas que confundían a los matemáticos, era el comportamiento de las series infinitas, sobre todo aquellas que no eran [convergentes](https://es.wikipedia.org/wiki/Serie_convergente). Recordemos que una serie es [convergente](https://es.wikipedia.org/wiki/Serie_convergente) cuando la suma de sus términos da como resultado un número finito; en caso contrario, se dice que la serie es [divergente](https://es.wikipedia.org/wiki/Serie_divergente). Así por ejemplo la [serie geométrica](https://es.wikipedia.org/wiki/Serie_geom%C3%A9trica), sabemos que converge en 1, y se puede demostrar geométricamente:

$$\sum\limits_{i=1}^\infty\frac{1}{2^n} = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \dots = 1$$

<img alt="demostracion serie geometrica" title="demostracion serie geometrica" src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e6/GeometricSquares.svg/359px-GeometricSquares.svg.png
" width="300px" height="300px">

Pero en cambio el resultado de la siguiente serie no se puede determinar:

$$ S = 1 - 1 + 1 - 1 + 1 - 1 + 1 - 1 + 1 \dots $$

Si por ejemplo agrupamos los términos de la siguiente forma, el resultado parece ser *obviamente* cero.

$$ S = (1 - 1) + (1 - 1) + (1 - 1) + (1 - 1) \dots $$

Pero si los agrupamos de esta otra forma, el resultado parece ser ahora *claramente* uno.

$$ S =  1 + (- 1 + 1) + (- 1 + 1) + (- 1 + 1) + (- 1 + 1) \dots $$

y no termina acá, ya que incluso podríamos agrupar la serie del siguiente modo y el resultado sería $\frac{1}{2}$.

$$ S = 1 - ( 1 - 1 + 1 - 1 + 1 - 1 + 1 - \dots) \Rightarrow S = 1 - S \Rightarrow 2S = 1 \Rightarrow S = \frac{1}{2} $$

es decir que el valor de la serie S, podría ser tanto 1, como 0, como $\frac{1}{2}$.No es sorprendente que argumentos como éste, hacían que los matemáticos se pusieran muy nerviosos al tratar con los *[infinitos](https://es.wikipedia.org/wiki/Infinito)* y quisieran evitarlos o incluso eliminarlos a toda costa.

Fuera de toda esta ambigüedad y confusión, la claridad surgió repentinamente en el siglo XIX, debido a los esfuerzos en solitario de un hombre brillante. [Georg Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) produjo una teoría que respondió a todas las objeciones de sus predecesores y reveló la riqueza inesperada escondida en el ámbito de lo *[infinito](https://es.wikipedia.org/wiki/Infinito)*. De repente, los [infinitos actuales](https://es.wikipedia.org/wiki/Infinito_potencial_e_infinito_actual) se convirtieron en parte de las [matemáticas](https://es.wikipedia.org/wiki/Matem%C3%A1ticas).

## Cantor y la teoría de conjuntos para explicar el infinito

[Georg Cantor](https://es.wikipedia.org/wiki/Georg_Cantor), aplicando ideas sumamente originales, postuló la [teoría de conjuntos](https://es.wikipedia.org/wiki/Teor%C3%ADa_de_conjuntos) y utilizando el concepto de la [cardinalidad](https://es.wikipedia.org/wiki/Conjunto#Cardinalidad) de los [conjuntos](https://es.wikipedia.org/wiki/Conjunto), se propuso explicar el concepto de *[infinito](https://es.wikipedia.org/wiki/Infinito)*.

Segun la teoría de [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor), dos [conjuntos](https://es.wikipedia.org/wiki/Conjunto) se dicen que tienen el mismo número de elementos, es decir, que tienen la misma [cardinalidad](https://es.wikipedia.org/wiki/Conjunto#Cardinalidad), si existe una función definida entre ellos de forma tal que a cada elemento le corresponde sólo otro elemento del otro [conjunto](https://es.wikipedia.org/wiki/Conjunto), y viceversa; o sea, que exista una correspondencia de uno a uno entre los elementos de ambos [conjuntos](https://es.wikipedia.org/wiki/Conjunto). A partir de esta definición se puede establecer la idea de [conjunto infinito](https://es.wikipedia.org/wiki/Conjunto_infinito). Se dice que un conjunto es *[infinito](https://es.wikipedia.org/wiki/Infinito)* si existe un subconjunto con la misma [cardinalidad](https://es.wikipedia.org/wiki/Conjunto#Cardinalidad) que él. Esta definición plantea una contradicción con la intuición, pues todo subconjunto como parte del conjunto total parece que deba tener menos elementos. Eso es así, efectivamente, en los [conjuntos finitos](https://es.wikipedia.org/wiki/Conjunto_finito), pero no en los *[infinitos](https://es.wikipedia.org/wiki/Infinito)*.

### ¿Distintos tipos de infinitos? Infinitos numerables y no numerables

Otro de los grandes aportes de [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor), fue la distinción entre los *[infinitos](https://es.wikipedia.org/wiki/Infinito)* [numerables](https://es.wikipedia.org/wiki/Conjunto_numerable) y [no numerables](https://es.wikipedia.org/wiki/Conjunto_no_numerable).
[Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) definió a los *[infinitos](https://es.wikipedia.org/wiki/Infinito)* [numerables](https://es.wikipedia.org/wiki/Conjunto_numerable) como aquellos en los que se puede encontrar una correspondencia uno-a-uno con la lista de [números naturales](https://es.wikipedia.org/wiki/N%C3%BAmero_natural) 1, 2, 3, 4, 5, 6,. . . Así, por ejemplo, los números pares son un *[infinito](https://es.wikipedia.org/wiki/Infinito)* [numerable](https://es.wikipedia.org/wiki/Conjunto_numerable), también lo son todos los números impares y ambos tienen el mismo tamaño. Por ejemplo, la correspondencia entre los primeros números impares sería la siguiente:

$$ 
1 \longrightarrow 3 \\
2 \longrightarrow 5 \\
3 \longrightarrow 7 \\
4 \longrightarrow 9 \\
5 \longrightarrow 11 \\
6 \longrightarrow 13 \\
7 \longrightarrow 15 \\
8 \longrightarrow 17 \\
9 \longrightarrow 19 \\
10 \longrightarrow \dots \\
$$

Por lo tanto, todos los [conjuntos infinitos numerables](https://es.wikipedia.org/wiki/Conjunto_numerable) tienen el mismo "tamaño" o [cardinalidad](https://es.wikipedia.org/wiki/Conjunto#Cardinalidad) en el sentido de [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor). El también pensaba que éstos eran los *[infinitos](https://es.wikipedia.org/wiki/Infinito)*  más pequeños que pudieran existir y por lo tanto los representó utilizando la primera letra del alfabeto hebreo, el símbolo [Aleph-cero](https://es.wikipedia.org/wiki/Alef_cero), $\aleph_{0}$. Un punto importante a tener en cuenta, es que esta definición excluye cualquier [conjunto finito](https://es.wikipedia.org/wiki/Conjunto_finito) de objetos, ya que un [conjunto finito](https://es.wikipedia.org/wiki/Conjunto_finito) sólo puede ponerse en correspondencia uno-a-uno con otro [conjunto](https://es.wikipedia.org/wiki/Conjunto) que contenga el mismo número de miembros.
Este análisis llevó a algunas conclusiones sorprendentes. [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) mostró que los [números racionales](https://es.wikipedia.org/wiki/N%C3%BAmero_racional), los cuales se forman al dividir un [número entero](https://es.wikipedia.org/wiki/Numeros_enteros) por otro (por ejemplo $\frac{1}{2}$, o $\frac{13}{2}$) son también un [conjunto infinito numerable](https://es.wikipedia.org/wiki/Conjunto_numerable). El truco fue encontrar un sistema para poder contarlos sin perderse ninguno. Él utilizó el famoso proceso ahora conocido como la [diagonal de Cantor](https://es.wikipedia.org/wiki/Argumento_de_la_diagonal_de_Cantor) para hacer esto. Fue contando cada uno de ellos de acuerdo al siguiente ordenamiento:

<img alt="Diagonal de Cantor" title="Diagonal de Cantor" src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/85/Diagonal_argument.svg/429px-Diagonal_argument.svg.png
" width="300px" height="300px">

[Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) también pudo demostrar con un nuevo tipo de argumento matemático que existen *[infinitos](https://es.wikipedia.org/wiki/Infinito)* más grandes y que no se pueden contar. Esto son los que hoy en día se conocen como [infinitos no numerables](https://es.wikipedia.org/wiki/Conjunto_no_numerable). Dentro de esta categoría podemos encontrar al conjunto de los [números reales](https://es.wikipedia.org/wiki/N%C3%BAmeros_reales), los cuales incluyen a los [números irracionales](https://es.wikipedia.org/wiki/N%C3%BAmero_irracional), quienes no se pueden escribir como [fracciones](https://es.wikipedia.org/wiki/Fracci%C3%B3n) y tienen una expansión infinita de sus parte decimal.

Para demostrar que los [números reales](https://es.wikipedia.org/wiki/N%C3%BAmeros_reales) son un [infinito no numerable](https://es.wikipedia.org/wiki/Conjunto_no_numerable), [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) comenzó por suponer que se podían contar, lo que significaba que debería ser capaz de elaborar una receta sistemática para contar todos los decimales interminables que no terminaran en una cadena infinita de ceros. Los primeros de estos números podrían parecerse a éstos:

$$
1 \longrightarrow 0.\underline{2}34567891\dots \\
2 \longrightarrow 0.5\underline{7}5603737\dots \\
3 \longrightarrow 0.46\underline{3}214516\dots \\
4 \longrightarrow 0.846\underline{2}16388\dots \\
5 \longrightarrow 0.5621\underline{9}4632\dots \\
6 \longrightarrow 0.46673\underline{2}271\dots \\
 \dots
$$

Luego creó un nuevo decimal tomando el primer dígito después del punto decimal del primer número, el segundo dígito del segundo número, y así sucesivamente para siempre(como por ejemplo los dígitos subrayados en la serie de arriba). Siguiendo el ejemplo, el nuevo decimal comenzaría como sigue:

$0.273292\dots$

Por último a este decimal recién creado, le sumó 1 a cada uno de sus infinitos dígitos. Obteniendo el número

$0.384303\dots$

El problema es que este número que acabó creando no puede aparecer en la lista ordenada original de todos los decimales que había asumido debía existir. Deberá ser distinto a cada número de la lista por lo menos en uno de sus dígitos, ya que fue construido expresamente de ese modo. Por lo tanto, los [números reales](https://es.wikipedia.org/wiki/N%C3%BAmeros_reales) (a veces llamados los números decimales o el continuo de números) no se pueden contar y son un [infinito no numerable](https://es.wikipedia.org/wiki/Conjunto_no_numerable). Así mismo, también deben ser más grande que los [números naturales](https://es.wikipedia.org/wiki/N%C3%BAmero_natural) o los [números racionales](https://es.wikipedia.org/wiki/N%C3%BAmero_racional). [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) representó a este nuevo tipo de *[infinito](https://es.wikipedia.org/wiki/Infinito)* con el símbolo $\aleph_{1}$; ya que creía que no debía existir otro tipo de *[infinito](https://es.wikipedia.org/wiki/Infinito)* que fuera mayor que el de los [números naturales](https://es.wikipedia.org/wiki/N%C3%BAmero_natural) y menor que el de los [números reales](https://es.wikipedia.org/wiki/N%C3%BAmeros_reales), aunque nunca fue capaz de demostrarlo. Esta es lo que se conoce como la [hipótisis del continuo](https://es.wikipedia.org/wiki/Hip%C3%B3tesis_del_continuo).

Este descubrimiento de [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor), que existen *[infinitos](https://es.wikipedia.org/wiki/Infinito)* de diferentes tamaños y se pueden distinguir de una manera completamente inequívoca, fue uno de los grandes descubrimientos de las [matemáticas](https://es.wikipedia.org/wiki/Matem%C3%A1ticas). También estaba completamente en contra de la opinión dominante de la época.

### Una jerarquía de infinitos

Además de todos los grandes aportes que [Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) realizó a la comprensión del *[infinito](https://es.wikipedia.org/wiki/Infinito)*, su descubrimiento más espectacular fue que los *[infinitos](https://es.wikipedia.org/wiki/Infinito)* no sólo son incontables, sino que son insuperables!. Descubrió que debe existir una jerarquía ascendente interminable de *[infinitos](https://es.wikipedia.org/wiki/Infinito)*. No hay uno más grande que todos que pueda contener a todos ellos. No hay un universo de universos que podemos anotar y capturar.
[Cantor](https://es.wikipedia.org/wiki/Georg_Cantor) fue capaz de demostrar que existe una jerarquía ascendente de *[infinitos](https://es.wikipedia.org/wiki/Infinito)* sin fin, una jerarquía infinita de *[infinitos](https://es.wikipedia.org/wiki/Infinito)*!. Para demostrar este punto, utilizó el concepto de [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia). 

Recordemos que un [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia) de un [conjunto](https://es.wikipedia.org/wiki/Conjunto) A, expresado por $P_{A}$, es el [conjunto](https://es.wikipedia.org/wiki/Conjunto) formado por todos los distintos [subconjuntos](https://es.wikipedia.org/wiki/Subconjunto) de A. Así por ejemplo el [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia) del [conjunto](https://es.wikipedia.org/wiki/Conjunto) $A=\{1,2,3\}$; va a ser igual a $P_{A}=\{\emptyset,\{1\},\{2\},\{3\},\{1,2\},\{2,3\}, \{1,3\},\{1,2,3\}\}$.
Un teorema importante de la [teoría de conjuntos](https://es.wikipedia.org/wiki/Teor%C3%ADa_de_conjuntos) establece que si A es un [conjunto](https://es.wikipedia.org/wiki/Conjunto) con k elementos, es decir que n(A) = k; entonces el [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia) de A tiene exactamente $2^k$ elementos. En nuestro ejemplo anterior podemos ver que n(A)=3, por lo tanto $n(P_{A}) = 2^3$, lo que es igual a los 8 elementos que vimos que tiene el [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia) de A.

Entonces, si se toma cualquier [conjunto infinito](https://es.wikipedia.org/wiki/Conjunto_infinito), siempre es posible generar una que es infinitamente más grande considerando a su [conjunto potencia](https://es.wikipedia.org/wiki/Conjunto_potencia), es decir, el conjunto que contiene todos sus subconjuntos. 
Así, de un [conjunto infinito](https://es.wikipedia.org/wiki/Conjunto_infinito) como $\aleph_{0}$ podemos crear un conjunto infinitamente más grande mediante la formación de su conjunto potencia, $P_{\aleph_{0}}$. Luego podemos hacer lo mismo otra vez, formando el conjunto potencia de $P_{\aleph_{0}}$, el cual será infinitamente más grande que $P_{\aleph_{0}}$. Y así sucesivamente, sin fin.
De esta forma las [matemáticas](https://es.wikipedia.org/wiki/Matem%C3%A1ticas) crean una jerarquía sin fin de *[infinitos](https://es.wikipedia.org/wiki/Infinito)* ascendentes . El *[infinito](https://es.wikipedia.org/wiki/Infinito)* nunca puede ser capturado por las fórmulas. También muestra que el número de posibles verdades es *[infinito](https://es.wikipedia.org/wiki/Infinito)*.


## Hotel infinito

Una de las imágenes que se suele utilizar para ilustrar el concepto de *[infinito](https://es.wikipedia.org/wiki/Infinito)* es el relato del [hotel infinito](https://es.wikipedia.org/wiki/El_hotel_infinito_de_Hilbert) ideado por el gran matemático [David Hilber](https://es.wikipedia.org/wiki/David_Hilbert), el cual explica de manera simple e intuitiva, las paradojas relacionadas con el concepto del *[infinito](https://es.wikipedia.org/wiki/Infinito)*.

El [hotel infinito](https://es.wikipedia.org/wiki/El_hotel_infinito_de_Hilbert) es un hotel como cualquier otro, con la única excepción de que cuenta con un número *[infinito](https://es.wikipedia.org/wiki/Infinito)* de habitaciones!

Tan pronto se abrieron las puertas de este hotel la gente comenzó a abarrotarlo y pronto se encontraron con que el hotel de habitaciones infinitas se encontraba lleno de infinitos huéspedes. En ese momento surgió la primera [paradoja](https://es.wikipedia.org/wiki/Paradoja), así que se tomó como medida que los huéspedes siempre tendrían habitación asegurada pero con el acuerdo previo de que tendrían que cambiar de habitación cada vez que se les pidiera.
Fue entonces cuando llegó un nuevo huesped al hotel que ya se encontraba lleno. El hombre pidió su habitación y el recepcionista, consciente de que no habría ningún problema, tomó un micrófono por el que avisó a todos los huéspedes que por favor revisaran el número de su habitación, le sumaran uno y se cambiaran a ese número de habitación, de esta manera el nuevo huésped pudo dormir tranquilamente en la habitación número 1. Pero, ¿qué pasó entonces con el huésped que se encontraba en la última habitación? Sencillamente no hay última habitación ya que $\infty + 1$ sigue siendo $\infty$.

La segunda [paradoja](https://es.wikipedia.org/wiki/Paradoja) surgió cuando llegó un representante de una agencia de viajes solicitando hospedar a un infinito número de turistas en el hotel de infinitas habitaciones, las cuales ya estaban todas ellas ocupadas. Pero el recepcionista nuevamente no tuvo ningún problema, tomó el micrófono y pidió a todos los huéspedes que se mudaran a la habitación correspondiente al resultado de multiplicar por 2 el número de su habitación actual. De esa forma todos los huéspedes se mudaron a una habitación par, y todas las habitaciones impares quedaron libres. Como hay infinitos números impares, los infinitos turistas pudieron alojarse sin más problemas.

La tercer [paradoja](https://es.wikipedia.org/wiki/Paradoja) se presentó cuando llegó otro representante de la agencia de viajes aún más preocupado que el primero y avisó que ahora la agencia tenía un infinito número de excursiones con un infinito número de turistas cada una. ¿cómo podrían hospedar a un número infinito de infinitos turistas en un hotel que ya se encontraba lleno?
El recepcionista permaneció inmutable, tomó tranquilamente el micrófono y se comunicó solamente con las habitaciones cuyo número fuera [primo](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) o alguna potencia de éstos ($p^n$), les pidió que elevaran el número 2 al número de la habitación en la que se encontraban ($2^{p^n}$) y se cambiaran a esa habitación.
Entonces asignó a cada una de las excursiones un [número primo](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) (distinto de 2), a cada uno de los turistas de cada una de las excursiones un número impar (t), de manera que la habitación de cada uno de los turistas, se calculaba tomando el [número primo](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) de su excursión (p) y elevandolo al número que les tocó dentro de su excursión (t) lo que da $p^t$.
Existiendo un número infinito de números primos y un número infinito de números impares, fácilmente se logró hospedar a un número infinito de infinitos huéspedes dentro del [hotel infinito](https://es.wikipedia.org/wiki/El_hotel_infinito_de_Hilbert).

## Las paradojas de Zenón y el concepto de límite

Como venimos viendo, el concepto de *[infinito](https://es.wikipedia.org/wiki/Infinito)* esta repleto de [paradojas](https://es.wikipedia.org/wiki/Paradoja); y algunas de las más famosas de ellas son también de las más antiguas. Se las debemos a  [Zenón de Elea](https://es.wikipedia.org/wiki/Zen%C3%B3n_de_Elea), quien fue discípulo del filósofo [Parménides](https://es.wikipedia.org/wiki/Parm%C3%A9nides_de_Elea) y por lo tanto sostenía, al igual que su maestro, que el universo era intemporal e inmutable y que por lo tanto el movimiento era solo una ilusión.
Para sostener su tesis, [Zenón de Elea](https://es.wikipedia.org/wiki/Zen%C3%B3n_de_Elea) propuso una serie de argumentos que hoy en día se conocen como las [paradojas de Zenón](https://es.wikipedia.org/wiki/Paradojas_de_Zen%C3%B3n).

La primera [paradoja](https://es.wikipedia.org/wiki/Paradoja) pretende demostrar que el movimiento es imposible, porque si quisiéramos caminar de un punto a otro, primero deberíamos cruzar la mitad de la distancia, luego la mitad de la distancia restante, luego la mitad del resto, y así sucesivamente. O sea, que si los dos puntos están a un kilómetro de distancia,  en primer lugar debemos llegar a $\frac{1}{2}$ km desde su inicio, a continuación, $\frac{3}{4}$ km desde su inicio, luego a $\frac{7}{8}$ km y así sucesivamente. Según el argumento de [Zenón](https://es.wikipedia.org/wiki/Zen%C3%B3n_de_Elea) sólo se podría llegar si se toma un número infinito de pasos; y como los griegos rechazaban el *[infinito](https://es.wikipedia.org/wiki/Infinito)*, por lo tanto deberíamos rechazar el movimiento también y concluir que este no es posible.

<img alt="Paradoja zenon" title="Paradoja de zenon" src="http://relopezbriega.github.io/images/zenon.png" width="500px" height="300px">

En su segunda [paradoja](https://es.wikipedia.org/wiki/Paradoja), [Zenón](https://es.wikipedia.org/wiki/Zen%C3%B3n_de_Elea) crea el escenario de una carrera entre el famoso atleta Aquiles y un rival mucho más lento, una tortuga. Aquiles comienza en la posición 0, mientras que la tortuga, que sólo puede correr a la mitad de la velocidad de Aquiles, tiene una ventaja inicial de un kilómetro. Ambos comienzan a correr al mismo tiempo. De acuerdo al planteamiento de este escenario, podríamos pensar que Aquiles, al correr el doble de rápido que la tortuga, la superaría en la marca de 2 kilómetros. Sin embargo, cuando Aquiles llega a la marca de  1 km, la tortuga ya ha avanzado a $1 + \frac{1}{2}$ km; cuando Aquiles llega al punto $1\frac{1}{2}$ km, la tortuga ha alcanzado $1 + \frac{1}{2} + \frac{1}{4}$ km; y así sucesivamente. Cuando, después de N pasos, Aquiles alcanza una distancia de $2 - \frac{1}{2}^{N-1}$ desde el punto inicial, la tortuga se encuentra todavía en la delantera, ya que está a una distancia de $2 - \frac{1}{2}^{N+1}$. No importa lo grande que sea N (el número de divisiones del viaje), Aquiles nunca se adelanta a la tortuga!

### Resolviendo las paradojas

Para poder resolver estas paradojas, debemos recurrir al [cálculo infinitesimal](https://es.wikipedia.org/wiki/C%C3%A1lculo_infinitesimal), y más particularmente al concepto de [límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico). El [límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) de una sucesión es la noción intuitiva de que la sucesión se aproxima arbitrariamente a un único punto o valor. Es decir, que la sucesión va a [converger](https://es.wikipedia.org/wiki/Serie_convergente) en un valor determinado.
En el caso de las dos [paradojas de Zenón](https://es.wikipedia.org/wiki/Paradojas_de_Zen%C3%B3n), nos encontramos ante [series geométricas](https://es.wikipedia.org/wiki/Serie_geom%C3%A9trica), las cuales ya vimos que tienen un [límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) y [convergen](https://es.wikipedia.org/wiki/Serie_convergente) en un valor finito.
Por tanto se puede demostrar que Aquiles realmente alcanzará a la tortuga. Los tiempos en los que Aquiles recorre la distancia que lo separa del punto anterior en el que se encontraba la tortuga son cada vez más y más pequeños (hasta el infinito más pequeños), y su suma da un resultado finito, que es el momento en que alcanzará a la tortuga.

## Límites al infinito

Como pudimos ver a lo largo de todo el artículo, el *[infinito](https://es.wikipedia.org/wiki/Infinito)* es un concepto muy especial. Sabemos que no podemos llegar a el, pero todavía podemos tratar de averiguar el valor de las funciones o sucesiones que tratan con el *[infinito](https://es.wikipedia.org/wiki/Infinito)*.
Por ejemplo, si quisiéramos saber cual es el resultado de $\frac{1}{\infty}$, deberíamos responder que no existe solución, ya que el *[infinito](https://es.wikipedia.org/wiki/Infinito)* no es un número, y por lo tanto no podríamos determinar el resultado de esa operación. 
Sin embargo, a pesar de que no podemos trabajar con el *[infinito](https://es.wikipedia.org/wiki/Infinito)* en sí mismo, podemos aproximarnos bastante a el y ver que sucede a medida que vamos dividiendo a uno por valores cada vez más grandes. Por ejemplo, si reemplazamos al $\infty$, por $x$, tendríamos una ecuación a la que podríamos ir aplicando valores cada vez más grandes de $x$ y ver a que resultado llegamos.
Ayudándonos de nuestro buen amigo [Python](http://python.org/) podríamos ver los resultados fácilmente y comprobar que a medida que $x$ se hace cada vez más grande, el valor de la función $\frac{1}{x}$ se aproxima cada vez más a cero.

```python
# <!-- collapse=True -->
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def f(x):
    return x**-1.0  # 1/x == x**-1

x = np.array([1, 2, 4, 10, 100, 1000, 10000, 100000, 1000000 ])
y = f(x)

# Construyendo tabla de valores x e y con pandas
tabla = pd.DataFrame(x, columns=['x'])
tabla['1/x'] = y
tabla
```

```python
# <!-- collapse=True -->
# Graficando 1/x
x = np.arange(1, 1000) # Rango de valores de x
plt.figure(figsize=(8,6))
plt.title(r"Graficando $\frac{1}{x}$")
plt.xlabel('x')
plt.ylabel(r"$f(x) = \frac{1}{x}$")
plt.plot(x, f(x), label=r"$\frac{1}{x}$ tiende a cero a medida que x crece")
plt.ylim([0, .12])
plt.legend()
plt.show()
```

Como nos muestran tanto los cálculos como el gráfico, a medida que vamos aumentando el valor de $x$, la función $\frac{1}{x}$ se va aproximando cada vez más a cero. Por tanto podemos conjeturar que la función $\frac{1}{x}$ tiende a cero.
Si bien, no podemos decir que pasa en el *[infinito](https://es.wikipedia.org/wiki/Infinito)*, si podemos decir que pasa a medida que nos vamos acercando cada vez más a él. Por tanto podríamos decir que a medida que el valor $x$ tiende a $\infty$, el valor de la función $\frac{1}{x}$, se aproxima a su [límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) cero. En el lenguaje de las [matemáticas](https://es.wikipedia.org/wiki/Matem%C3%A1ticas), lo expresaríamos de la siguiente forma:

$$\lim_{x\to \infty} \left(\frac{1}{x}\right) = 0$$

El concepto de [Límite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) es una de las herramientas principales sobre las que se sustenta el [cálculo](https://es.wikipedia.org/wiki/C%C3%A1lculo_infinitesimal). Muchas veces, una función puede ser indefinida en un punto, pero podemos pensar en lo que pasa a medida que la función "se acerca" cada vez más a ese punto. Otras veces, la función puede ser definida en un punto, pero puede aproximarse a un límite diferente.
